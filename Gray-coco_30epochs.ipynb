```python
%cd /
from glob import glob

t_img_list = glob('/home/mate0329/dataset/train/images/*.jpg')

print(len(t_img_list))
```

```python
v_img_list = glob('/home/mate0329/dataset/valid/images/*.jpg')

print(len(v_img_list))
```

### 트레이닝 이미지 경로, 벨리데이션 이미지 경로 txt 파일로 저장

```python
with open('/home/mate0329/dataset/train.txt', 'w') as f:
  f.write('\n'.join(t_img_list) + '\n')

with open('/home/mate0329/dataset/val.txt', 'w') as f:
  f.write('\n'.join(v_img_list) + '\n')
```

### data.yaml 수정 (train,validation)

```python
import yaml

with open('/home/mate0329/dataset/data.yaml', 'r') as f:
  data = yaml.load(f)

print(data)

data['train'] = '/home/mate0329/dataset/train.txt'
data['val'] = '/home/mate0329/dataset/val.txt'

with open('/home/mate0329/dataset/data.yaml', 'w') as f:
  yaml.dump(data, f)

print(data)
```
